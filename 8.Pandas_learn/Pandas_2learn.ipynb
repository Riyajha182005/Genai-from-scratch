```python
## Specifying columns data types
data1=('a,b,c,d\n'
 '1,2,3,4\n'
 '5,6,7,8\n'
 '9,10,11,12')
```

```python
print(data1)
```

```python
df=pd.read_csv(StringIO(data1),dtype=object)
```

```python
df
```

```python
df['a'][1] ## giving me a string datatype for sure
```

```python
## Lets convert into another datatype
df=pd.read_csv(StringIO(data1),dtype=int)
```

```python
df
```

```python
df['a']
```

```python
df=pd.read_csv(StringIO(data1),dtype={'a':int,'b':float,'c':object,'d':int})
```

```python
df['c']
```

```python
## check the datatypes
df.dtypes
```

```python
## Index columnn and training delimiters
data2=('index,a,b,c\n'
'4,apple,bat,5.7\n'
'8,orange,ball,10')
```

```python
pd.read_csv(StringIO(data2))
```

```python
## Making the first column is my index
pd.read_csv(StringIO(data2),index_col=0)
```

```python
data3=('a,b,c\n'
'4,apple,bat,\n'
'8,orange,ball,')
```

```python
pd.read_csv(StringIO(data3))
```

```python
pd.read_csv(StringIO(data3),index_col=False)
```

```python
## Combining usecols and index_col
pd.read_csv(StringIO(data3),usecols=['b','c'],index_col=False)
```

```python
## Quoting and escape characters .....very useful in NLP
data4='a,b\n"hello, \\"Bob\\", nice to meet you",5'
```

```python
pd.read_csv(StringIO(data4),escapechar='\\')
```

```python
pd.read_csv(StringIO(data4),escapechar='\\',quotechar='"')
```
